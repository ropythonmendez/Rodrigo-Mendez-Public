```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent="myGeocoder")
location = geolocator.geocode(loc)
location
```

```python
geolocator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=0.1)
```

```python
df = (pd.read_csv("col_vendor.csv")
      # rename column
      .rename(columns={'Point A': 'point_a'})
      # Geocode address and then extract lat, lng values
      .assign(raw_coords = lambda df: np.vectorize(geocode)(df.point_a),
              longitude = lambda df: df.raw_coords.apply(lambda x: x.longitude if x else None),
              latitude = lambda df: df.raw_coords.apply(lambda x: x.latitude if x else None)
             )
     )
```

```python
df.head()
```

```python
df.info()
```

```python
df = (df
      # rename column
      .rename(columns={'Point B': 'point_b'})
      # Geocode address and then extract lat, lng values
      .assign(raw_coords_dest = lambda df: np.vectorize(geocode)(df.point_b),
              dest_longitude = lambda df: df.raw_coords_dest.apply(lambda x: x.longitude if x else None),
              dest_latitude = lambda df: df.raw_coords_dest.apply(lambda x: x.latitude if x else None)
             )
     )
```

```python
df.info()
```

```python
geo_df = (df.drop(['DoW','Vendor','Campaign_name','Hour_1', 'point_a', 'point_b', 'raw_coords', 'raw_coords_dest','Year','Week' ],axis =1
                  , errors = 'ignore')
         )
geo_df.head()
```

```python

geo_df = (geo_df
          .rename(columns={'Start Date': 'date'})
         )
```

```python
geo_df['date'] = geo_df.date.astype('datetime64[ns]')
```

```python
geo_df = geo_df.assign (date = geo_df['date'].dt.strftime('20%y/%m/%d'),
                       year = geo_df['date'].dt.strftime('20%y'),
                       weekday = geo_df['date'].dt.strftime('%w'),
                        week = geo_df['date'].dt.strftime('%W'),
                       Mins = lambda x: round(x.Mins),
                        Km = lambda x: round(x.Km)
                       )
```
